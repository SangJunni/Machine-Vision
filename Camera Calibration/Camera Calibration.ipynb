```python
import numpy as np
import cv2 as cv
import glob
from google.colab.patches import cv_imshow
```

```python
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)

objp = np.zeros((6*8, 3), np.float32)
objp[:,:2]= np.mgrid[0:8,0:6].T.reshape(-1,2)

objpoints = []
imgpoints = []
images = glob.glob('/content/gdrive/MyDrive/MachineVision/cb*.jpg')
```

```python
i=0
for fname in images:
  print(fname)
  img =cv.imread(fname)
  img = cv.resize(img,(int(800),int(600)))
  #cv_imshow(gray)
  gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
  ret, corners = cv.findChessboardCorners(gray,(8,6), None)
  print(ret)
  if ret ==True:
    objpoints.append(objp)
    corners2 = cv.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)
    imgpoints.append(corners)

    cv.drawChessboardCorners(img,(8,6), corners2, ret)
    cv_imshow(img)
    cv.imwrite('/content/gdrive/MyDrive/MachineVision/HW1_result/result'+str(i)+'.jpg',img)
    i=i+1
```

```python
ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None, None)
```

```python
def draw(img,corners, imgpts):
  imgpts = np.int32(imgpts).reshape(-1,2)

  img = cv.drawContours(img, [imgpts[:4]], -1, (0,255,0),-3)

  for i,j in zip(range(4),range(4,8)):
    img = cv.line(img,tuple(imgpts[i]), tuple(imgpts[j]),(255,0,0),3)
  
  img = cv.drawContours(img, [imgpts[4:]], -1, (0,0,255), 3)
  return img
```

```python
axis = np.float32([[0,0,0],[0,3,0],[1,3,0],[1,0,0],[0,0,-5],[0,3,-5], [1,3,-5],[1,0,-5]])

i=0
for fname in glob.glob('/content/gdrive/MyDrive/MachineVision//cb*.jpg'):
  img = cv.imread(fname)
  img = cv.resize(img,(int(800),int(600)))
  gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
  ret, corners = cv.findChessboardCorners(gray,(8,6), None)
  #print(ret)
  if ret == True:
    imgpts, jac = cv.projectPoints(axis, rvecs[i],tvecs[i], mtx, dist)
    i = i+1
    img = draw(img,corners2, imgpts)
    cv_imshow(img)
    cv.imwrite('/content/gdrive/MyDrive/MachineVision/HW1_result2/result'+str(i)+'.jpg',img)
```
